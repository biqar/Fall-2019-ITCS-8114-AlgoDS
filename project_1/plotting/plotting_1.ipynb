```python
# importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
r_df = pd.read_fwf('execution_time_random_io.txt', header=None)
s_df = pd.read_fwf('execution_time_sorted_io.txt', header=None)
rs_df = pd.read_fwf('execution_time_rev_sorted_io.txt', header=None)
```

```python
# numpy representation
# random
r_m1=r_df.values
r_m2=r_m1.reshape((5,30))
r_m3=np.zeros((5,10))
for j in range(10):
    r_m3[:,j]= np.mean(r_m2[:,j*3:(j+1)*3], axis=1)
r_m3
# sorted
s_m1=s_df.values
s_m2=s_m1.reshape((5,30))
s_m3=np.zeros((5,10))
for j in range(10):
    s_m3[:,j]= np.mean(s_m2[:,j*3:(j+1)*3], axis=1)
s_m3
# rev sorted
rs_m1=rs_df.values
rs_m2=rs_m1.reshape((5,30))
rs_m3=np.zeros((5,10))
for j in range(10):
    rs_m3[:,j]= np.mean(rs_m2[:,j*3:(j+1)*3], axis=1)
rs_m3
```

```python
# plotting
list_is = [5,10,15,20,25,30,35,40,45,50]

# random
fig1 = plt.figure(figsize=(12,8))
plt.clf()
plt.title('Input-size vs Time (random)')
plt.plot(list_is, r_m3[0,:].reshape(-1,1), '*-', color='red', label="Insertion-sort")
plt.plot(list_is, r_m3[1,:].reshape(-1,1), 'x-', color='green', label="Merge-sort")
plt.plot(list_is, r_m3[2,:].reshape(-1,1), '+-', color='blue', label="Heapsort")
plt.plot(list_is, r_m3[3,:].reshape(-1,1), 'g^-', color='purple', label="In-place quicksort")
plt.plot(list_is, r_m3[4,:].reshape(-1,1), 'bo-', color='cyan', label="Modified quicksort")
plt.yscale('log')
plt.ylabel('Time')
plt.xlabel('No of samples')
plt.legend(loc='best')
fig1.tight_layout()
```

```python
# # sorted
fig2 = plt.figure(figsize=(12,8))
plt.clf()
plt.title('Input-size vs Time (sorted)')
plt.plot(list_is, s_m3[0,:].reshape(-1,1), '*-', color='red', label="Insertion-sort")
plt.plot(list_is, s_m3[1,:].reshape(-1,1), 'x-', color='green', label="Merge-sort")
plt.plot(list_is, s_m3[2,:].reshape(-1,1), '+-', color='blue', label="Heapsort")
plt.plot(list_is, s_m3[3,:].reshape(-1,1), 'g^-', color='purple', label="In-place quicksort")
plt.plot(list_is, s_m3[4,:].reshape(-1,1), 'bo-', color='cyan', label="Modified quicksort")
plt.yscale('log')
plt.ylabel('Time')
plt.xlabel('No of samples')
plt.legend(loc='best')
fig2.tight_layout()
```

```python
# # rev sorted
fig3 = plt.figure(figsize=(12,8))
plt.clf()
plt.title('Input-size vs Time (rev sorted)')
plt.plot(list_is, rs_m3[0,:].reshape(-1,1), '*-', color='red', label="Insertion-sort")
plt.plot(list_is, rs_m3[1,:].reshape(-1,1), 'x-', color='green', label="Merge-sort")
plt.plot(list_is, rs_m3[2,:].reshape(-1,1), '+-', color='blue', label="Heapsort")
plt.plot(list_is, rs_m3[3,:].reshape(-1,1), 'g^-', color='purple', label="In-place quicksort")
plt.plot(list_is, rs_m3[4,:].reshape(-1,1), 'bo-', color='cyan', label="Modified quicksort")
plt.yscale('log')
plt.ylabel('Time')
plt.xlabel('No of samples')
plt.legend(loc='best')
fig3.tight_layout()
```
